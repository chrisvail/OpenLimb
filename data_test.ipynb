```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\chris\Downloads\1D_limb_data.csv")
data
```

```python
measurements = data[['W1', 'W2', 'W3', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'L1', 'L2', 'L3']]
measurements.isna().sum(0)
```

Calculate the number of nan free rows in the dataset.

```python
(measurements.isna().sum(1) == 0).sum()
```

```python
# Create a binary dataframe indicating NaN values (1 for NaN, 0 for non-NaN)
nan_indicators = measurements.isna().astype(int)

# Calculate the correlation matrix of NaN patterns
nan_correlation = nan_indicators.corr()

sns.heatmap(nan_correlation, square=True)
```
